# **🗨️ [문제1] 결측치 유무 파악하기**

```python
df.isnull().sum()
```

이상치 유무 파악하기

```python
import numpy as np
def outliers_iqr(dt, col):
    q1, q3 = np.percentile(dt[col], [25, 75])
    iqr = q3 - q1
    outlier_bottom = q1 - (1.5 * iqr)
    outlier_top = q3 + (1.5 * iqr)
    outliers_1 = dt[col] < outlier_bottom
    outliers_2 = dt[col] > outlier_top
    outliers = dt[outliers_1 | outliers_2]
    return outliers[[col]]
find_outliers = outliers_iqr(df, 'Fare')
find_outliers
```

# **🗨️ [문제2] 불필요한 데이터 속성 삭제하기**

```python
df_core_columns = df.drop(['PassengerId', 'Name', 'Ticket'], axis = 1)
```

# **🗨️ [문제3] 데이터의 결측치가 많은 ‘Cabin’ 속성을 제거하기**

```python
df_notnull = df_core_columns.drop(['Cabin'], axis = 1)
```

# **🗨️ [문제4] 결측치가 있는 데이터 제거**

```python
df_notnull = df_notnull.dropna()
```

### ⭐ 이상치가 있는 데이터를 치환하기



# **🗨️ [문제5] 'Fare' 속성의 이상치를 결측치로 변경하고 확인**

```python
find_outliers = outliers_iqr(df_notnull, 'Fare')
df_notnull.loc[find_outliers.index, 'Fare'] = np.nan
df_notnull.isnull().sum()
```

## **🗨️ [문제6] 이상치 치환하기**
(이상치 → 결측치로 변경 후 평균값으로 치환)

```python
mean_Fare = df_notnull['Fare'].mean(axis = 0)
df_notnull['Fare'].fillna(df_notnull['Fare'].mean(), inplace = True)
df_notnull.loc[find_outliers.index, 'Fare']
```
