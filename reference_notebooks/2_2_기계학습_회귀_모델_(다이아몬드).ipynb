seaborn 라이브러리는 다이아몬드의 여러 특성에 대한 데이터셋 diamonds를 제공한다.

속성들은 각각 캐럿, 컷팅 유형, 색상, 투명도, 깊이, 테이블 사이즈, 가격, 가로 길이, 세로 길이, 높이를 뜻한다.

데이터셋을 활용해 다이아몬드 가격을 예측하는 지도학습 모델을 구현해보자!

📍 [데이터 설명]
다이아몬드 데이터 설명.png

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
df = sns.load_dataset('diamonds')

# 혹시,, 런타임에 연결할 수 없습니다. 로 뜬다면? 런타임 → 런타임 유형 변경 → None (CPU) 확인!
```

```python
df.head()
```

```python
df.info()
```

```python
x = df[['x','y','z']] #모델이 예측의 힌트로 사용할 속성
y = df['price'] #모델이 예측하고자 하는 정답 속성
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression() #사용할 모델 라이브러리 불러오기
model = LinearRegression() #선택한 기계학습 알고리즘의 객체 생성

model.fit(x_train, y_train) #훈련 데이터를 입력하여 모델 학습시키기
```

```python
from sklearn.metrics import r2_score #성능 지표 라이브러리 불러오기
Predicted = model.predict(x_test)
r2_score(y_test, Predicted) #성능 지표 값(결정계수) 출력
```
