```python
df.dropna(subset=['num1'], inplace=True) # 특정 열(num1)의 결측치가 있는 행만 삭제
df.head()
```

**③ 평균값, 최빈값으로 대체하기**
→ 다시 일부 데이터 값을 none처리 필요

```python
import seaborn
df = seaborn.load_dataset('anagrams')

df.loc[[2,3],['num1']] = None # 일부 데이터를 결측치(None)로 제작
df.loc[[1,3],['num2']] = None
df.loc[3,['attnr']] = None
df.head()
print('\n결측치 개수:')
print(df.isnull().sum())
```

```python
mean_num1=df['num1'].mean() #num1 열의 평균
df['num1']=df['num1'].fillna(mean_num1) # 결측치 채우기
df.head()
```

```python
mode_attnr=df['attnr'].mode()[0] #attnr 열의 최빈값
df['attnr']=df['attnr'].fillna(mode_attnr) # 결측치 채우기
df.head()
```

```python
df.loc[[0,2,3],['attnr']] = 'divided*' # 데이터 값 수정
df['attnr'].head()
```

**④ 텍스트값 대체하기**

```python
df['attnr'] = df['attnr'].str.replace("*", "", regex=False) # * 문자를 공백으로 처리
df['attnr'].head()
```
